```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data pelanggan/content/WhatsgoodlyData-7.csv
df = pd.read_csv('/content/Live_20210128_facebook.csv')

# Memilih atribut yang akan digunakan untuk klasterisasi
#X = df[['num_reactions', 'num_likes']]
X = df[['num_comments', 'num_likes']]
# Mengubah data kategorikal menjadi numerik menggunakan one-hot encoding
X = pd.get_dummies(X)

# Menentukan jumlah klaster yang diinginkan
kmeans = KMeans(n_clusters=3)

# Melakukan klasterisasi
kmeans.fit(X)

# Menambahkan kolom klaster ke dalam data pelanggan
df['Cluster'] = kmeans.labels_

# Visualisasi hasil klasterisasi
#plt.scatter(df['num_reactions'], df['num_likes'], c=df['Cluster'])
plt.scatter(df['num_comments'], df['num_likes'], c=df['Cluster'])
#plt.xlabel('num_reactions')
plt.xlabel('num_comments')
plt.ylabel('num_likes')
plt.show()
```

```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data pelanggan
df = pd.read_csv('/content/Live_20210128_facebook.csv')

# Memilih atribut yang akan digunakan untuk klasterisasi
X = df[['num_reactions', 'num_likes']]

# Mengubah data kategorikal menjadi numerik menggunakan one-hot encoding
X = pd.get_dummies(X)

# Menentukan jumlah klaster yang diinginkan
kmeans = KMeans(n_clusters=3)

# Melakukan klasterisasi
kmeans.fit(X)

# Menambahkan kolom klaster ke dalam data pelanggan
df['Cluster'] = kmeans.labels_

# Visualisasi hasil klasterisasi
plt.scatter(df['num_reactions'], df['num_likes'], c=df['Cluster'])
plt.xlabel('snum_reactions')
plt.ylabel('num_likes')
plt.show()
```
